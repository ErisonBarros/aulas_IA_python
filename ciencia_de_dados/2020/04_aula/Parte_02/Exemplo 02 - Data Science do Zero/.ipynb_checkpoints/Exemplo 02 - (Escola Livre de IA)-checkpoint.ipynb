Solu√ß√£o Generalista - atrav√©s de M√©todos Num√©ricos:


Solu√ß√£o Anal√≠tica - Par√¢metros possuem F√≥rmulas Definidas:<br>
(V√°lido para o caso espec√≠fico da Regress√£o Linear)



Usando a Solu√ß√£o Anal√≠tica:

```python
import numpy as np
def solucao_analitica(x,y):
    beta = np.corrcoef(x, y)[0,1] * np.std(y) / np.std(x)
    alpha = mean(y) - beta * mean(x)
    return alpha, beta

alpha, beta = solucao_analitica(x, y)
print("alpha", alpha)
print("beta", beta)
```

Usando a Solu√ß√£o Generalista:





Anima√ß√£o Dispon√≠vel em:
https://github.com/pvigier/gradient-descent/blob/master/README.md

```python
def y_i_estim(alfa, beta, x_i):
    return beta * x_i + alfa

def erro_i(alfa, beta, x_i, y_i):
    return y_i - y_i_estim(alfa, beta, x_i)

def quadrado_erro_i(alfa, beta, x_i, y_i):
    return erro_i(alfa, beta, x_i, y_i) ** 2

def soma_quadrado_erros(alfa, beta, x, y):
    return sum(erro_i(alfa, beta, x_i, y_i) ** 2
               for x_i, y_i in zip(x, y))

def gradiente_quadrado_erros(x_i, y_i, theta):
    alfa, beta = teta
    return [-2 * erro_i(alfa, beta, x_i, y_i),        # Dire√ß√£o de Ajuste em Alfa
            -2 * erro_i(alfa, beta, x_i, y_i) * x_i]  # Dire√ß√£o de Ajuste em Beta
```

#### Par√¢metros do Gradiente Descendente

```python
learning_rate = 0.01
initial_b = 0
initial_m = 0
num_iterations = 300
```

#### Fun√ß√£o Custo - fun√ß√£o a ser minimizada

```python
def compute_cost(b, m, points):
    total_cost = 0
    N = float(len(points))
    
    #Compute sum of squared errors
    for i in range(0, len(points)):
        x = points[i, 0]
        y = points[i, 1]
        total_cost += (y - (m * x + b)) ** 2
        
    #Return average of squared error
    return total_cost/N
```

#### C√°lculo do Gradiente Descendente

```python
def gradient_descent_runner(points, starting_b, starting_m, learning_rate, num_iterations):
    b = starting_b
    m = starting_m
    cost_graph = []

    #For every iteration, optimize b, m and compute its cost
    for i in range(num_iterations):
        cost_graph.append(compute_cost(b, m, points))
        b, m = step_gradient(b, m, array(points), learning_rate)

    return [b, m, cost_graph]

def step_gradient(b_current, m_current, points, learning_rate):
    m_gradient = 0
    b_gradient = 0
    N = float(len(points))

    #Calculate Gradient
    for i in range(0, len(points)):
        x = points[i, 0]
        y = points[i, 1]
        m_gradient += - (2/N) * x * (y - (m_current * x + b_current))
        b_gradient += - (2/N) * (y - (m_current * x + b_current))
    
    #Update current m and b
    m_updated = m_current - learning_rate * m_gradient
    b_updated = b_current - learning_rate * b_gradient

    #Return updated parameters
    return b_updated, m_updated
```

#### Finalmente: Obten√ß√£o dos Par√¢metros a e b

```python
b, m, cost_graph = gradient_descent_runner(points, initial_b, initial_m, learning_rate, num_iterations)

#Print optimized parameters
print ('Optimized b:', b)
print ('Optimized m:', m)

#Print error with optimized parameters
print ('Minimized cost:', compute_cost(b, m, points))
```

#### O Movimento do Gradiente Descendente

```python
plt.plot(cost_graph)
plt.xlabel('No. of iterations')
plt.ylabel('Cost')
plt.title('Cost per iteration')
plt.show()
```

Gradient descent converges to local minimum after 5 iterations

#### Plot line of best fit

```python
#Plot dataset
plt.scatter(x, y)
#Predict y values
pred_grad = m * x + b
pred_formula = beta * x + alpha
#Plot predictions as line of best fit
plt.plot(x, pred_grad, c='r')
plt.plot(x, pred_formula, c='k')
plt.xlabel('Quantidade de Amigos')
plt.ylabel('Tempo na Rede Social')
plt.title('Compara√ß√£o das Retas de Regress√£o')
plt.show()
```

### An√°lise do Modelo Final

```python
import numpy as np

correlation_matrix = np.corrcoef(x, y)

#print(correlation_matrix)

correlation_xy = correlation_matrix[0,1]

print("Correlacao_XY = ", round(correlation_xy, 4))

r_squared = correlation_xy**2
print("R^2 = ", round(r_squared, 4))
```

### Considera√ß√µes Finais

O valor de R^2 obtido nesse exemplo apresenta um valor ainda menor. Isso significa que o nosso modelo de √∫nica vari√°vel independente est√° deixando muitas informa√ß√µes fora do radar, deixando a previs√£o cada vez menos assertiva.
