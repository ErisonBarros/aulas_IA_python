### Análise dos Data Sets

    - Obtenção dos Dados
    - Plot Inicial: primeiras intuições
---

```python
from numpy import *
import matplotlib.pyplot as plt

points = genfromtxt('data.csv', delimiter=',')

#Extract columns
x = array(points[:,0])
y = array(points[:,1])

#Plot the dataset
plt.scatter(x,y)
plt.xlabel('Horas de Estudo')
plt.ylabel('Notas na Prova')
plt.title('Conjunto de Dados - Intuição')
plt.show()
```

---
### Análise da Curva de Regressão

    - Obtenção dos Parâmetros
    - Plot da Reta de melhor Ajuste
    - Realização de Previsões
---

Solução Generalista - através de Métodos Numéricos:


Solução Analítica - Parâmetros possuem Fórmulas Definidas:<br>
(Válido para o caso específico da Regressão Linear)



```python
m = (len(x)*cov(x, y)[0,1])/((len(x) - 1)*var(x))
#m = np.cov(x, y)[0,1]/np.var(x)

b = mean(y) - m * mean(x)

print(m)
print(b)
```

```python
#b = 0.02963934787473239
#m = 1.4774173755483797

#Plot dataset
plt.scatter(x, y)
#Predict y values
pred = m * x + b
#pred1 = m1 * x + b1
#Plot predictions as line of best fit
plt.plot(x, pred, c='r')
#plt.plot(x, pred1, c='r')
plt.xlabel('Horas de Estudo')
plt.ylabel('Notas na Prova')
plt.title('Conjunto de Dados - Reta de Regressão')
plt.show()
```

#### Previsões através da Curva de Regressão

Vamos selecionar 4 pontos do nosso Data Set e fazermos previsões.

```python
#b =  0.02963934787473239
#m = 1.4774173755483797

x_reais = [42.43647694, 52.21179669, 55.14218841, 61.53035803]
y_reais = [76.61734128, 79.64197305, 78.21151827, 62.5623823]

y_regressao =  [m * x + b for x in x_reais]

plt.scatter(x_reais, y_reais, c='b')
plt.scatter(x_reais, y_regressao, c='k')

pred = m * x + b
plt.plot(x, pred, c='r')

plt.show()

erro_proporcional = [round(100*(1 - i/j),2) for i, j in zip(y_reais, y_regressao)]
print(erro_proporcional)
```

---
### Análise do Modelo Final
    - Cálculo da Correlação
    - Pontos de Melhoria do Modelo
---

```python
import numpy as np

correlation_matrix = np.corrcoef(x, y)

print(correlation_matrix)

correlation_xy = correlation_matrix[0,1]

print("Correlacao_XY = ", round(correlation_xy, 4))

r_squared = correlation_xy**2
print("R^2 = ", round(r_squared, 4))
```

### Considerações Finais

Não deve ser uma surpresa o fato de encontrarmos um coeficiente de correlação considerado alto de acordo com a nossa tabela de referência.

Nos exemplos didáticos, é comum encontrarmos correlações altas, acima de 0.70.

Entretanto, nos casos reais, esses valores são muito difíceis de acontecer, pois são bastante óbvios e diretos, e em geral estamos buscando compreender problemas complexos e obter insights que ainda não foram descobertos.

---

Além disso, devemos sempre tentar compreender o que está acontecendo com o modelo e buscar formas de melhorar os valores previstos.

Poderíamos adicionar ao cálculo das Notas na Prova não somente as Horas de Estudo, mas também a Frequência nas Aulas, ou a Existência de um Material mais focado, como provas antigas.

Esse modelo de regressão linear que recebe como parâmetro uma série de variáveis independentes é chamado de Regressão Múltipla, que será apresentada na próxima Aula da Escola Livre de Inteligência Artificial.
